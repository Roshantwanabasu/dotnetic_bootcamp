```python
import pandas as pd
import numpy as np
```

```python
def linear_regression(X,Y,theta,epochs=100,learning_rate = 0.1):
    for iter1 in range(epochs):
        y_current = np.dot(X,theta)
        theta -= (learning_rate/len(X))*np.sum(np.dot((y_current-Y).T , X))
    cost =  1/len(X)*np.sum(np.power((y_current-Y),2))
    return theta,cost
```

```python
df = pd.read_excel("house_price.xlsx")
x = df['area'].values
x = x.astype(float)
x = x/x.max()
x = x[:,np.newaxis]
y = df['price'].values
y = y[:,np.newaxis]
y = y.astype(float)
y = y/y.max()
ones=np.ones((41,1))
x = np.concatenate((ones,x),axis=1)
theta = np.random.random((x.shape[1],1))
theta, error = linear_regression(x,y,theta)
print(theta,error)
```
